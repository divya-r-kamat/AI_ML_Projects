## Visualization

```python
sns.set(color_codes=True)
```

```python
#Drop na values
games.dropna(inplace=True)
```

```python
#Distance plot for minage and average rating
sns.distplot(games['minage'])
```

```python
sns.distplot(games['average_rating'])
```

#### Is there a linear relation ship between minage and average_rating

```python
sns.jointplot(games['minage'],games['average_rating'])
```

#### Compare the relationship between playing time, minage and average rating

```python
sns.pairplot(games[['playingtime','minage','average_rating']])
```

#### Compare type of game and playingtime using a strip plot

```python
sns.stripplot(games['type'],games['playingtime'],jitter=True)
```

```python
sns.regplot('playingtime','average_rating',data=games[games['playingtime']<500])
```
